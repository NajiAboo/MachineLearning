```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/NajiAboo/MachineLearning/main/datasets/titanic.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
data = df[['Survived', 'Pclass', 'Sex','SibSp','Parch','Fare','Age']]
```

```python
data.head()
```

```python
data['Sex']= [ 1 if x == 'male' else 0for x in data['Sex']]
```

```python
data.head()
```

```python
test_data = data[data['Age'].isnull()]
```

```python
test_data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
X_train = data.drop('Age', axis=1)
```

```python
X_test = test_data.drop('Age', axis=1)
```

```python
y_train = data['Age']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
data_with_age = X_test.copy()
```

```python
data_with_age['Age'] = y_pred
```
